## Building dataframes for integration with pytorch geometric

```python
all_interacting_drugs = pd.concat((drugcomb_scored["Drug1"], drugcomb_scored["Drug2"])) # concat drug1 and drug2
all_interacting_drugs = all_interacting_drugs[(all_interacting_drugs.duplicated() == False)] # drop duplicates
all_interacting_drugs
```

```python
# Drop interacting drugs that do not have information (are not in the drug_chemical_info df)
all_interacting_drugs.isin(list(drug_chemical_info.index))
all_interacting_drugs = all_interacting_drugs[all_interacting_drugs.isin(list(drug_chemical_info.index))]

# Convert to dataframe
interacting_drugs_infos = pd.DataFrame(all_interacting_drugs, columns=['name'])

# Add drug infos
all_interacting_drugs.apply(lambda s: drug_chemical_info.at[str(s), 'smilesString'])
interacting_drugs_infos['smiles'] = all_interacting_drugs.apply(lambda s: 
                                                                drug_chemical_info.at[str(s), 'smilesString'])
interacting_drugs_infos['cIds'] = all_interacting_drugs.apply(lambda s: 
                                                                drug_chemical_info.at[str(s), 'cIds'])
```

```python
interacting_drugs_infos['smiles']
```

```python
# Retrieve all protein names
all_proteins = set(protein_protein_interactions['protein1']).\
union(set(protein_protein_interactions['protein2'])).union(set(drug_protein_link['protein']))
```
